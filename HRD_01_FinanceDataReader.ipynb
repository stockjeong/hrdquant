### 2. 증시 데이터 수집

#### 주식 데이터

```python
KP_stock_list = fdr.StockListing("KOSPI") # 코스피 종목 목록 불러오기
display(KP_stock_list) # 불러온 목록 보기
```

```python
sp_data = fdr.DataReader("005380", "2025-01-01", "2025-05-07") # 2025년 1월 1일부터, 5월 7일까지의 현대자동차(005380) 주식 데이터 불러오기
display(sp_data)
```

```python
SP500_stock_list = fdr.StockListing("SP500") # S&P 500 종목 불러오기 (https://ko.wikipedia.org/wiki/S%26P_500)
display(SP500_stock_list)
```

```python
sp_data = fdr.DataReader("AAPL", "2025-01-01", "2025-05-07") # 2025년 1월 1일부터, 5월 7일까지의 애플(AAPL) 주식 데이터 불러오기
display(sp_data)
```

#### 주요 지수 데이터

```python
KP_idx_data = fdr.DataReader("KS11", "2020-01-01", "2025-05-07") # 해당 기간의 코스피 idx 지수 불러오기
display(KP_idx_data)
```

```python
DJI_idx_data = fdr.DataReader("DJI", "2025-01-01", "2025-05-07") # 해당 기간의 다우지수 idx 불러오기
display(DJI_idx_data)
```
